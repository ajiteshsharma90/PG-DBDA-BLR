```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("rain.csv")
df.head()
```

```python
log1 = df[['pressure','rain']]
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(df['pressure']).reshape(-1,1), df['rain'])
```

```python
model.coef_[0]
```

```python
model.intercept_
```

```python
z = [model.coef_[0] * val + model.intercept_ for val in df['pressure']]
```

```python
def to_sigmoid(z):
    y = 1/(1+(np.exp(-z)))
    return y
```

```python
z_sig = [to_sigmoid(val) for val in z ]
```

```python
z_sig = np.array(z_sig)
z_sig
```

```python
yhat = np.where(z_sig>0.5,1,0)
```

```python
yhat
```
