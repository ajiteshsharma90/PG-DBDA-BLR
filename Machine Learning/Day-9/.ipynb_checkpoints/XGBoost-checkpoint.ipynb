```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
from xgboost import XGBClassifier
```

```python
df = pd.read_csv(r"C:\Users\dbda\Desktop\Suchandra & Ajitesh\Machine Learning\Day-9\income.csv")
df
```

```python
df.columns
```

```python
df.info()
```

```python
x = df[['age','education.num','marital.status','capital.gain', 'capital.loss', 'hours.per.week']]
x = pd.get_dummies(x)
y = df['income']
y = y.map({'<=50K':0,'>50K':1})
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size = 0.30)
```

```python
help(XGBClassifier)
```

```python
model = XGBClassifier()
```

```python
model.fit(x_train,y_train)
```

```python
ypred = model.predict(x_test)
score = accuracy_score(y_test,ypred)
print(score)
```

```python
linmod = LogisticRegression(max_iter=800)
linmod.fit(x_train,y_train)
ypredl = linmod.predict(x_test)
scorel = accuracy_score(y_test,ypredl)
print(scorel)
```
