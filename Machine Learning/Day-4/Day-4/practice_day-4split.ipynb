```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
```

```python
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25)
```

```python
train_scores = []
test_scores = []
```

```python
for i in range(1,10):
    x= np.array(df['topspeed']).reshape(-1,1)
    y= np.array(df['fuel_economy'])
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.25,random_state = (42+i))
    model=LinearRegression()
    model.fit(x_train,y_train)
    training_score = model.score(x_train, y_train)
    testing_score = model.score(x_test, y_test)
    train_scores.append(training_score)
    test_scores.append(testing_score)
```

```python
plt.plot(range(1,10),train_scores,label = 'Training Scores')
plt.plot(range(1,10),test_scores,label = 'Testing Scores')
plt.legend()
plt.show()
```
