```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score , r2_score, confusion_matrix, classification_report
```

```python
df = pd.read_excel('TASK_TWO.xlsx')
df.head()
```

```python
sns.pairplot(df)
plt.show()
```

```python
sns.heatmap(df.corr(), annot = True)
plt.show()
```

```python
tr_score = []
te_score = []
```

```python
for i in range(1,50):
    x = df[['water','mud']]
    y = df['yield']
    x_train,x_test,y_train,y_test = train_test_split(x,y,test_size= 0.25,random_state=(50+i)) 
    mod = LinearRegression()
    mod.fit(x_train,y_train)
    training_score = mod.score(x_train, y_train)
    testing_score = mod.score(x_test, y_test)
    tr_score.append(training_score)
    te_score.append(testing_score)
```

```python
plt.plot(range(1,50),tr_score, label = 'Training Score')
plt.plot(range(1,50),te_score, label = 'Testing Score')
plt.legend()
plt.show()
```

```python
dft = pd.DataFrame({'No': range(1,50),'Train': tr_score,'test': te_score})
no = dft.loc[dft['test'] == max(dft['test']),'No']
no
```
