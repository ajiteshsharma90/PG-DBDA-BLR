```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("data47.csv")
df
```

```python
df.replace(["c", "a", "b"], np.nan, inplace = True)
df
```

```python
df["num1"] = df["num1"].fillna(df["num1"].median())
df["num2"] = df["num2"].fillna(df["num2"].median())
df
```

```python
df.info()
```

```python
df=df.astype(float)
df.info()
```

```python
df["sum"] = df["num1"] + df["num2"]
df["average"] = df["sum"] / 2
df
```
