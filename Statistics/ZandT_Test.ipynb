Hypothesis Test

1. Parametric Z-Test

```python
import warnings
warnings.filterwarnings('ignore')
import pandas as pd
#Loading data
passport_df = pd.read_csv('passport.csv')

passport_df.head(5)
```

```python
#Printing all the records
print(list(passport_df.processing_time))
```

```python
#Conducting Z-test for this hypothesis
import math
from scipy import stats

def z_test( pop_mean, pop_var, sample ):
    z_score = (sample.mean() - pop_mean)/(pop_var/math.sqrt(len(sample)))
    return z_score, stats.norm.cdf(z_score)
```

```python
#Calling function for conducting hypothesis
z_test( 30, 12.5, passport_df.processing_time )
```

```python
#the return is (Z,p)
#Since 6.77% probability is more than 5%
#Hence hypothesis is accepted
```

T-Test

1.One-Sample T-Test

```python
#Loading bolloywood movie data
bollywood_movies_df = pd.read_csv( 'bollywoodmovies.csv' )
```

```python
#displaying 5 samples
bollywood_movies_df.head(5)
```

```python
#Printing the list of samples of production cost
print(list(bollywood_movies_df.production_cost))
```

```python
#Conducting 1 sample t test
stats.ttest_1samp( bollywood_movies_df.production_cost, 500 )
```

```python
#the return is (T-statistics,p)
#Since 2.786% probability is less than 5%
#Hence hypothesis is rejected
```

2.Two-Sample-Test

```python
#In this test two sample means are avaible, SD is not
```

```python
#Load healthy drinking data, with yes drink people
healthdrink_yes_df = pd.read_excel( 'healthdrink.xlsx', 'healthdrink_yes')
healthdrink_yes_df.head(5)
```

```python
#Load healthy drinking data, with no drink people
healthdrink_no_df = pd.read_excel( 'healthdrink.xlsx', 'healthdrink_no')
healthdrink_no_df.head(5)
```

```python
#See the distribution of data with yes and no option
import seaborn as sn
import matplotlib.pyplot as plt
sn.distplot( healthdrink_yes_df['height_increase'], label ='healthdrink_yes' )
sn.distplot( healthdrink_no_df['height_increase'], label ='healthdrink_no' )
plt.legend();
```

```python
#The distribution of increase in ht. of those drink as
#compare to no drink
#But is this enough to claim; lets see
```

```python
#Conducting Two-Sample T-test
stats.ttest_ind(healthdrink_yes_df['height_increase'],
                healthdrink_no_df['height_increase'])
```

```python
#the return is (Tstatistics,p)
#Since 0% probability is less than 5%
#Hence hypothesis is rejected
```
